```python
# Import libraries

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
plt.style.use('seaborn')
```

```python
# Load Messi's data 
Messi = pd.ExcelFile("C:\Users\Michael\Documents\Python Projects\Messi v CR7\Messi Data (Not Copa Del Rey).xlsx")
Messi = pd.read_excel("C:\Users\Michael\Documents\Python Projects\Messi v CR7\Messi Data (Not Copa Del Rey).xlsx")
Messi
```

```python
# Drop data corresponing to Argentina matches
Messi = Messi[Messi.Team !='Argentina']
Messi
```

```python
# Drop data for the 2018/19 season 
Messi = Messi.drop([0])
Messi = Messi.drop([1])
Messi
```

```python
# Rename Columns 
Messi = Messi.rename(columns={'Apps': 'Appearances', 'Mins' : 'Minutes Played', 'Yel': 'Yellow Cards', 'Red': 'Red Cards', 'SpG' : 'Shots Per Game','PS%' : 'Passing Accuracy', 'AerialsWon': 'Aerials Won', 'MotM': 'Man of the Match awards'})
Messi
```

```python
# Drop the 'Team' column
Messi = Messi.drop(Messi.columns[1], axis=1)
Messi
```

```python
# Drop the 'Tournament' column
Messi = Messi.drop(Messi.columns[1], axis=1)
# Fill in the blanks
Messi = Messi.fillna(0)
Messi
```

```python
# Group the rows by season
Messi = Messi.groupby('Season', as_index=False).agg({'Appearances' : 'sum', 'Minutes Played': 'sum', 'Goals':'sum', 'Assists': 'sum', 'Yellow Cards' : 'sum', 'Red Cards': 'sum', 'Shots Per Game' : 'mean', 'Passing Accuracy': 'mean', 'Aerials Won': 'mean', 'Man of the Match awards': 'sum', 'Rating' :'mean'}).reindex(columns=Messi.columns)
Messi
```

```python
# Load Messi's Copa Del Rey Data
Messi_Cup_Data = pd.ExcelFile("C:\Users\Michael\Documents\Python Projects\Messi v CR7\Messi Cup Data.xlsx")
Messi_Cup_Data = pd.read_excel("C:\Users\Michael\Documents\Python Projects\Messi v CR7\Messi Cup Data.xlsx")
Messi_Cup_Data
```

```python
# Rename teh 'Season' column
Messi_Cup_Data = Messi_Cup_Data.rename(columns = {'Season' : 'Year'})
Messi_Cup_Data
```

```python
Messi_Cup_Data
```

```python
# Concatenate Messi with Messi_Cup_Data
Messi_Total = pd.concat([Messi, Messi_Cup_Data], axis=1, join='outer')
Messi_Total
```

```python
# Combine the stats from Messi and Messi_Cup_Data
Total_Appearances = Messi_Total['Appearances'] + Messi_Total['M']
Messi_Total['Total Appearances'] = Total_Appearances

Total_Mintues_Played = Messi_Total['Minutes Played'] + Messi_Total['Min']
Messi_Total['Total Mintues Played'] = Total_Mintues_Played

Total_Goals = Messi_Total['Goals'] + Messi_Total['G']
Messi_Total['Total Goals'] = Total_Goals

Total_Assists = Messi_Total['Assists'] + Messi_Total['A']
Messi_Total['Total Assists'] = Total_Assists

Total_Yellow_Cards = Messi_Total['Yellow Cards'] + Messi_Total['YC']
Messi_Total['Total Yellow Cards'] = Total_Yellow_Cards

Total_Red_Cards = Messi_Total['Red Cards'] + Messi_Total['RC']
Messi_Total['Total Red Cards'] = Total_Red_Cards

Messi_Total
```

```python
# Drop unneccessary columns 
Messi_Total = Messi_Total.drop(Messi_Total.columns[1:7], axis=1)
Messi_Total
```

```python
Messi_Total = Messi_Total.drop(Messi_Total.columns[7:13], axis=1)
Messi_Total
```

```python
Messi_Total = Messi_Total.drop(Messi_Total.columns[6], axis=1)
Messi_Total
```
